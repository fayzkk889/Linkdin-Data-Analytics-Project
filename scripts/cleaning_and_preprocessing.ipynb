```python
import pandas as pd 
import numpy as np
```

```python
linkdin_data= pd.read_csv("linkedin_data_jobs_with_salary.csv")
linkdin_data.head()
```

```python
linkdin_data.drop("scraped_at", axis=1, inplace=True)
```

```python
linkdin_data.tail()
```

```python
linkdin_data.info()
```

```python
linkdin_data.isna().sum()
```

```python
linkdin_data['posted_date'].isnull().sum() # checking whether there's a null value in the posted_date column
```

```python
linkdin_data['posted_date'] = pd.to_datetime(linkdin_data['posted_date'], errors='coerce') #converted to date time for trend analysis
linkdin_data["posted_date"].info
```

```python
linkdin_data = linkdin_data.drop_duplicates()
linkdin_data["posted_date"].info #data has no duplicates
```

```python
#making the strings lowercase for consistencies
linkdin_data['title'] = linkdin_data['title'].str.lower().str.strip()
linkdin_data['company'] = linkdin_data['company'].str.lower().str.strip()
linkdin_data['location'] = linkdin_data['location'].str.lower().str.strip()
```
